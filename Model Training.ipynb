```python
import pandas as pd
import re
```

```python
features = pd.read_csv("kddcup.names", sep="\t", header=None, skiprows=1)
```

```python
features
```

```python
extract_names = re.compile('[A-Za-z_]{1,}')
```

```python
features_list = []

for i in range(features.shape[0]):
    ft = extract_names.findall(str(features.iloc[i]))
    features_list.append(ft[0])
```

```python
features_list.append('Label')
```

```python
features_list
```

```python
data = pd.read_csv("kddcup.data.corrected", names=features_list, header=None)

test_data = pd.read_csv("test_labeled_corrected", names = features_list, header = None)
```

```python
test_data
```

```python
data
```

```python
y_train = data["Label"]
```

```python
y_train
```

```python
y_test = test_data["Label"]
```

```python
y_test
```

```python
x_train = data.drop('Label', axis=1)
```

```python
x_train
```

```python
x_test = test_data.drop('Label', axis=1)
```

```python
x_test
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for i in x_train.columns:
    if str(x_train[i].dtype) == 'object':
        x_train[i] = encoder.fit_transform(x_train[i])
        x_test[i] = encoder.fit_transform(x_test[i])
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
scaled_x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=x_test.columns)
```
